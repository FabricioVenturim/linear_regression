# Variáveis

## Editando o dataframe com as informações do brasileirão

```python
# public = pd.read_csv("estadiosdefutebol.csv")
championship_data = pd.read_csv("campeonato-brasileiro-full.csv")
```

```python
championship_data
```

```python
filtered_values = np.where(((championship_data["mandante"] == "Flamengo") | (championship_data["visitante"] == "Flamengo")) & (championship_data["data"] >= "2016-01-01") & (championship_data["data"] <= "2020-01-01"))
flamengo_games = championship_data.loc[filtered_values]
flamengo_games.reset_index(inplace=True)
flamengo_games.drop(axis=1, inplace = True, labels = ["formacao_mandante", "formacao_visitante", "tecnico_mandante", "tecnico_visitante", "arena","mandante_estado","visitante_estado","estado_vencedor", "data", "mandante_placar", "visitante_placar"])
flamengo_games
```

```python
filtered_values = np.where((flamengo_games["mandante"] == "Flamengo"))
flamengo_home_games = flamengo_games.loc[filtered_values]
flamengo_home_games
```

## Variável "Dia"

```python
list_days = []

for game in flamengo_home_games.index:
    if flamengo_home_games["dia"][game] in ["Sabado", "Domingo"]:
        list_days.append(1)
    else:
        list_days.append(0)
flamengo_home_games.drop(axis=1, inplace = True, labels = ["dia"])     
flamengo_home_games["dia_nobre"] = list_days
```

## Horario

Horário   | Final de semana | Dia útil
--------- | ------ |---------
[11:00,18:00) | Cedo | Cedo
[18:00, 19:30) | Médio | Cedo
[19:30, 20:30) | Tarde | Cedo
[20:30, 21:50) | Tarde | Médio
[21:50, 22:30) | Tarde | Tarde

Horário   | HB1 | HB2
--------- | ------| -------
Cedo | 0 | 1
Médio | 1 | 0
Tarde | 1 | 1

```python
HB1 = []
HB2 = []


for game in flamengo_home_games.index:
    if (((flamengo_home_games["dia_nobre"][game] == 1) and ("11:00" <= flamengo_home_games["hora"][game] < "18:00")) or ((flamengo_home_games["dia_nobre"][game] == 0) and ("16:00" <= flamengo_home_games["hora"][game] < "20:30"))):
        HB1.append(0)
        HB2.append(1)
    elif (((flamengo_home_games["dia_nobre"][game] == 1) and ("18:00" <= flamengo_home_games["hora"][game] < "19:30")) or ((flamengo_home_games["dia_nobre"][game] == 0) and ("20:30" <= flamengo_home_games["hora"][game] < "21:50"))):
        HB1.append(1)
        HB2.append(0)
    elif (((flamengo_home_games["dia_nobre"][game] == 1) and ("19:30" <= flamengo_home_games["hora"][game] < "22:30")) or ((flamengo_home_games["dia_nobre"][game] == 0) and ("21:50" <= flamengo_home_games["hora"][game] < "22:30"))):
        HB1.append(1)
        HB2.append(1)

flamengo_home_games.drop(axis=1, inplace = True, labels = ["hora"])   
flamengo_home_games["HB1"] = HB1
flamengo_home_games["HB2"] = HB2
            
```

```python
flamengo_home_games
```

```python
R1 = []
R2 = []
R3 = []
R4 = []
list_games = [R1,R2,R3,R4]
for game in flamengo_home_games.index:
    for list_game, index in zip(list_games, range(game-1, game-5, -1)):
        if index < 0:
            list_game.append(np.nan)
        elif flamengo_games["vencedor"][index] == "-":
            list_game.append(1)
        elif flamengo_games["vencedor"][index] == "Flamengo":
            list_game.append(3)
        else:
            list_game.append(0)
            
flamengo_home_games["R1"] = R1
flamengo_home_games["R2"] = R2
flamengo_home_games["R3"] = R3
flamengo_home_games["R4"] = R4
```

```python
flamengo_home_games
```

```python
local_rivals = ["Botafogo-RJ", "Vasco", "Fluminense"]
national_rivals = ["Atletico-MG", "Corinthians", "Cruzeiro", "Gremio", "Internacional", "Palmeiras", "Santos", "Sao Paulo"]

list_rivalry_1 = []
list_rivalry_2 = []


for game in flamengo_home_games.index:
    if flamengo_home_games["visitante"][game] in local_rivals:
        list_rivalry_1.append(1)
        list_rivalry_2.append(1)
    elif flamengo_home_games["visitante"][game] in national_rivals:
        list_rivalry_1.append(0)
        list_rivalry_2.append(1)
    else:
        list_rivalry_1.append(0)
        list_rivalry_2.append(0)

flamengo_home_games.drop(axis=1, inplace = True, labels = ["mandante", "visitante", "vencedor", "ID", "index"])   
        
        
flamengo_home_games["list_rivalry_1"] = list_rivalry_1
flamengo_home_games["list_rivalry_2"] = list_rivalry_2
```
